## PROGRAMACIÓN ORIENTADA A OBJETOS

En **Python**, todos los datos son **Objetos**, desde las funciones hasta las estructuras de datos. La programación orientada a objetos (POO) es una manera de programar (paradigma) que permite llevar al código mecanismos usados con entidades de la vida real.

Sus beneficios son los siguientes:

* **Encapsulamiento**:
Permite ***Empaquetar*** el código dentro de una unidad (**Objeto**) donde se puede determinar el ámbito de actuación.

* **Abstracción**:
Permite ***Generalizar*** los tipos de **Objetos** a través de las **Clases** y simplificar el programa.

* **Herencia**:
Permite ***Reutilizar*** código al poder heredar atributos y comportamientos de una **Clase** a otra.

* **Polimorfismo**:
Permite ***Crear*** múltiples **Objetos** a partir de una misma pieza flexible de código.

#### ¿Qué es un Objeto?
Un **Objeto** es una estructura de datos personalizada que contiene datos y código:

| Elementos | ¿Qué son? | ¿Cómo se llaman? | ¿Cómo se identifican? |
|---|---|---|---|
| Datos  | Variables | Atributos | Mediante sustantivos |
| Código  | Funciones | Métodos | Mediante verbos |

Un **Objeto** representa una instancia única de alguna entidad (a través de los valores de sus atributos) e interactúa con otros **Objetos** (o consigo mismo) a través de sus métodos.

<center><img src="./Imagenes/Auto.png" style="width:670px;height:300px;" class="center"></center>

#### ¿Qué es una Clase?
Para crear un **Objeto** primero debemos definir la **Clase** que lo contiene. Podemos pensar en la **Clase** como el molde con el que se crean nuevos **Objetos** de ese tipo, y puede haber tantas **Clases** como sean necesarias. Una vez que se crea un **Objeto** específico de una **Clase** se le considera como una instancia de esa **Clase**.

<center><img src="./Imagenes/Auto_C_O.png" style="width:816px;height:300px;" class="center"></center>

Las instancias o las **Clases** pueden guardar información, lo que se denomina **Atributos**. Por otro lado, cada **Clase** puede definir herramientas para operar tanto con sus propios datos como con otros **Objetos**, a las cuales se les denomina como **Métodos** y pueden ser de distintos tipos.

Por ejemplo, para la **Clase** ***Automóvil***, tres **Objetos** que serán sus instancias pueden ser ***Volkswagen Vento***, ***Fiat Panda*** y ***Chevrolet Malibu***.

### Creando Objetos

Se creará una primera **Clase** donde se van a modelar algunos automóviles como se describió previamente. Para ello se utilizará la palabra reservada `class` seguida del nombre de la **Clase**.

La documentación sobre las **Clases** en **Python** se puede consultar [en esta liga](https://docs.python.org/3/tutorial/classes.html).

***Nota:*** Los nombres de las **Clases** suelen ser escritos en formato ***camelCase*** (esta notación combina las palabras directamente, sin usar ningún símbolo, estableciendo que la primera letra de cada palabra esté en mayúscula a excepción de la primera palabra, estando el resto de letras en minúsculas) y en singular. Adicionalmente, la palabra reservada `pass` se emplea como un ubicador de futuro código, es decir, cuando se ejecuta no sucede nada pero se evita el generar un error por falta de código escrito.

```python
#Creación de la Clase
class Automovil:
    pass
```

Existen multitud de automóviles, pero una vez que se ha definido la **Clase** genérica es posible crear **Objetos** (instancias) concretos. 

En este caso y considerando los automóviles ***Volkswagen Vento***, ***Fiat Panda*** y ***Chevrolet Malibu***, la inicialización de cada instancia se realiza de la siguiente manera, y adicionalmente se verifica el tipo de dato contenido en cada una de las variables por medio de la función `type()`.

```python
#Creación de los Objetos
vw_vento = Automovil()
ft_panda = Automovil()
ch_malib = Automovil()

#Impresión de los tipos de Clase de cada Objeto
print("El Volkswagen Vento es un objeto del tipo:", type(vw_vento))
print("El Fiat Panda es un objeto del tipo:      ", type(ft_panda))
print("El Chevrolet Malibú es un objeto del tipo:", type(ch_malib))
```

### Añadiendo Métodos

Un **Método** es una función que forma parte de una **Clase** o de un **Objeto**. En su ámbito tiene acceso a otros métodos y atributos de la **Clase** o del **Objeto** al que pertenece.

La definición de un **Método** (de instancia) es análoga a la de una función ordinaria, pero incorporando un primer parámetro `self` que hace referencia a la instancia actual del **Objeto**.

Una de las acciones más sencillas que se pueden hacer sobre un automóvil es encenderlo o apagarlo, por lo mismo se implementarán estos dos métodos en la **Clase** (`.encendido()` y `.apagado()`):

```python
#Creación de la Clase
class Automovil:
    def encendido(self):
        print("El encendido del Automóvil ha sido correcto...")

    def apagado(self):
        print("Apagando el Automóvil...")

#Creación del Objeto
vw_vento = Automovil()

#Encendido y Apagado del Automóvil
vw_vento.encendido()
vw_vento.apagado()
```

### Añadiendo Atributos

Los **Atributos** son variables presentes en **Objetos** o en **Clases**, y se encargan de guardar la información. Pueden guardar no solo valores, sino también funciones asignadas tras la creación de la instancia u otros **Objetos** de otro tipo. 

Suponiendo que se quiere guardar en un **Atributo** el estado del automóvil (encendido/apagado), se puede realizar de la siguiente manera.

***Nota:*** Siempre que se quiera acceder a cualquier método o atributo del objeto habrá que utilizar la palabra reservada `self`, cuyo objetivo es hacer referencia al **Objeto** que se está manipulando cuando se llama al **Método**.

```python
#Creación de la Clase
class Automovil:
    def encendido(self):
        self.estado_on = True
        print("El encendido del Automóvil ha sido correcto...")

    def apagado(self):
        self.estado_on = False
        print("Apagando el Automóvil...")

#Creación del Objeto
vw_vento = Automovil()

#Arrancado del Automóvil y verificación de su estado
vw_vento.encendido()
print("¿El Automóvil está encendido?:", vw_vento.estado_on, "\n")

#Apagado del Automóvil y verificación de su estado
vw_vento.apagado()
print("¿El Automóvil está encendido?:", vw_vento.estado_on, "\n")
```

### Inicialización

Existe un método especial que se ejecuta cuando se crea una instancia de un **Objeto**. Este método es `__init__` y permite asignar **Atributos** y realizar operaciones con el **Objeto** en el momento de su creación. También se le conoce como el ***Constructor***.

Por ejemplo, se guardará el nombre del automóvil como un **Atributo** del **Objeto**:

```python
#Creación de la Clase
class Automovil:
    def __init__(self, nombre):
        self.nombre = nombre

#Creación del Objeto con su nombre
vw_vento = Automovil("Volkswagen Vento")

#Impresión del nombre del Automóvil
print("El nombre del Automóvil es:", vw_vento.nombre)
```

Aunque el **Atributo** correspondiente al nombre se ha creado en el ***Constructor*** de la **Clase**, también es posible modificarlo de la siguiente manera:

```python
#Asignación del nombre al Objeto
vw_vento.nombre = 'Vento'

#Impresión del nombre del Automóvil
print("El nombre del Automóvil es:", vw_vento.nombre)
```

Ahora se agregarán nuevos **Atributos**:
* Nombre del automóvil (`nombre`).
* Marca del automóvil (`marca`).
* Año del modelo (`modelo`).
* Tamaño del motor (`motor`).

Adicionalmente, se agregarán dos **Métodos**:
* Un **Método** llamado `.edadAuto()`para determinar la edad del automóvil, para ello se importará la librería **DateTime** por medio de `import datetime` para con ello poder definir la fecha actual. La edad se calcula simplemente con la diferencia entre el año actual y el año del modelo.
* Un **Método** llamado `.mostrarInfo()` el cual mostrará la información del automóvil en base a los valores de los **Atributos**.

Finalmente, se definirán tres **Objetos** correspondientes a los automóviles ***Volkswagen Vento***, ***Fiat Panda*** y ***Chevrolet Malibu*** con valores correspondientes a sus respectivos **Atributos**. De esa manera se podrá aplicar cada uno de los dos **Métodos** a los tres **Objetos** definidos. 

***Nota:*** En el **Método** `.edadAuto()` se puede definir la variable `edad` de manera global para que pueda ser usada fuera de la **Clase**, para ello se define la variable de esa manera por medio de la palabra reservada `global`.

***Observación:*** Podrá notarse la importancia del uso de la palabra reservada `self` cuyo objetivo es hacer referencia al **Objeto** que se está manipulando cuando se llama al **Método**. Por ejemplo, cuando se invoca el **Método** `.mostrarInfo()` para el **Objeto** correspondiente al ***Chevrolet Malibu***, `self` hace referencia a los valores de los **Atributos** de este **Objeto** y no a los valores de alguno de los otros dos **Objetos**.

```python
#Importación de librería y determinación de la fecha actual
import datetime
today = datetime.date.today()
year = today.year

#Creación de la Clase
class Automovil:
    def __init__(self, nombre, marca, modelo, motor):
        self.nombre = nombre
        self.marca  = marca
        self.modelo = modelo
        self.motor  = motor

    def edadAuto(self):
        #global edad
        edad = year - self.modelo
        print("El automóvil %s tiene %d años en el mercado." %(self.nombre, edad))
        
    def mostrarInfo(self):
        print("El automóvil %s es de la marca %s, el modelo %d tiene un motor de %.1f litros." 
              %(self.nombre, self.marca, self.modelo, self.motor))
        
#Creación de los Objetos con sus respectivos Atributos 
vw_vento = Automovil('Vento', 'Volkswagen', 2024, 2.0)
ft_panda = Automovil('Panda', 'Fiat', 2023, 1.8)
ch_malib = Automovil('Malibu', 'Chevrolet', 2020, 3.5)

#Aplicación de los Métodos para el primer Objeto
vw_vento.mostrarInfo()
vw_vento.edadAuto()
print("\n")

#Aplicación de los Métodos para el segundo Objeto
ft_panda.mostrarInfo()
ft_panda.edadAuto()
print("\n")

#Aplicación de los Métodos para el tercer Objeto
ch_malib.mostrarInfo()
ch_malib.edadAuto()
```

Ahora se agregarán nuevos **Atributos** quedando de la siguiente manera:
* Nombre del automóvil (`nombre`).
* Marca del automóvil (`marca`).
* Año del modelo (`modelo`).
* Tamaño del motor (`motor`).
* Tamaño de los neumáticos (`llanta`).
* Kilometraje actual (`odometro`).

Adicionalmente, se agregarán dos **Métodos** quedando de la siguiente manera:
* Un **Método** llamado `.edadAuto()`para determinar la edad del automóvil, para ello se importará la librería **DateTime** por medio de `import datetime` para con ello poder definir la fecha actual. La edad se calcula simplemente con la diferencia entre el año actual y el año del modelo.
* Un **Método** llamado `.mostrarInfo()` el cual mostrará la información del automóvil en base a los valores de los **Atributos**.
* Un **Método** llamado `.datosLlanta()` el cual mostrará el detalle de los neumáticos del automóvil en base a la nomenclatura estándar de su medida.
* Un **Método** llamado `.tasaKm()` el cual calculará la cantidad de kilómetros que el automóvil ha recorrido por año. Para automóviles del mismo año, se agrega una función `if` para evitar una divisón entre 0.

Con ello, se definirá un **Objeto** correspondiente a un ***Volkswagen Vento*** con valores específicos para sus **Atributos**. De esa manera se podrán aplicar los dos **Métodos** nuevos que se agregaron.

```python
#Creación de la Clase
class Automovil:
    def __init__(self, nombre, marca, modelo, motor, llanta, odometro):
        self.nombre   = nombre
        self.marca    = marca
        self.modelo   = modelo
        self.motor    = motor
        self.llanta   = llanta
        self.odometro = odometro

    def edadAuto(self):
        edad = year - self.modelo
        print("El automóvil %s tiene %d años en el mercado." %(self.nombre, edad))

    def mostrarInfo(self):
        print("El automóvil %s es de la marca %s, el modelo %d tiene un motor de %.1f litros." 
              %(self.nombre, self.marca, self.modelo, self.motor))
        
    def datosLlanta(self):
        print("Los datos de las llantas del automóvil %s son:" %self.nombre)
        print("Ancho de la llanta [milímetros]: %s" %self.llanta[0:3])
        print("Perfil de la llanta [porcentaje]: %s" %self.llanta[4:6])
        print("Diámetro del rin [pulgadas]: %s" %self.llanta[7:9])
        
    def tasaKm(self):
        if (year - self.modelo) == 0:
            tasa = self.odometro
        else:
            tasa = self.odometro / (year - self.modelo)
        print("El automóvil %s ha recorrido %.2f kilómetros por año." %(self.nombre, tasa))
        
#Creación del Objeto con sus respectivos Atributos
vw_vento = Automovil('Vento', 'Volkswagen', 2020, 2.0, '185/60R15', 80000)

#Aplicación del Método para el Objeto
vw_vento.datosLlanta()
print("\n")

#Aplicación del Método para el Objeto
vw_vento.tasaKm()
```

El uso de **Clases** permite tener múltiples instancias distintas de la misma **Clase**. Estas se comportan de la forma que se ha definido en la **Clase**, e incluso posibilitan generar **Clases** más específicas de la **Clase** original (denominada ***padre***), a esto se le conoce como **Herencia**.

Supongamos que se desea una nueva **Clase** para automóviles deportivos, sin embargo los **Atributos** y los **Métodos** que contiene la **Clase** `Automovil` también serán empleados en estos nuevos **Objetos**, pero contendrán ciertos valores específicos propios de un automóvil deportivo. 

Para evitar el duplicar una **Clase** y aprovechar al máximo lo que ya se tiene, se puede considerar como una **Clase** ***padre*** a `Automovil` y **Heredar** a una **Clase** ***hijo*** a la que se le llamará `automovilDeportivo` la cual heredará los **Atributos** y los **Métodos** pero a la vez tendrá los suyos propios.

Se va a considerar un ***McLaren 750S*** modelo 2020 el cual tiene un motor de 4.0 litros y alcanza una velocidad máxima de 332 kilómetros por hora empleando sus 750 caballos de potencia, y el cual tiene un peso de 1,277 kilogramos.

<center><img src="./Imagenes/McLaren.png" style="width:485px;height:200px;" class="center"></center>

Para ello se va a generar una nueva **Clase** ***hijo*** de nombre `automovilDeportivo` la cual heredará de la **Clase** ***padre*** `Automovil`. Los **Atributos** son los mismos de la **Clase** ***padre*** pero se agregarán tres nuevos, quedando de la siguiente manera:
* Nombre del automóvil (`nombre`), es un **Atributo** heredado.
* Marca del automóvil (`marca`), es un **Atributo** heredado.
* Año del modelo (`modelo`), es un **Atributo** heredado.
* Tamaño del motor (`motor`), es un **Atributo** heredado.
* Tamaño de los neumáticos (`llanta`), es un **Atributo** heredado.
* Kilometraje actual (`odometro`), es un **Atributo** heredado. 
* Velocidad máxima (`vel_maxima`), es un **Atributo** nuevo.
* Potencia (`potencia`), es un **Atributo** nuevo.
* Peso (`peso`), es un **Atributo** nuevo.

De igual manera, se definen dos nuevos **Métodos** que se agregan a los que se heredaron de la **Clase** ***padre*** quedando de la siguiente manera:
* Un **Método** llamado `.edadAuto()`, es un **Método** heredado.
* Un **Método** llamado `.mostrarInfo()`, es un **Método** heredado.
* Un **Método** llamado `.datosLlanta()`, es un **Método** heredado.
* Un **Método** llamado `.tasaKm()`, es un **Método** heredado.
* Un **Método** llamado `.mostrarInfoDep()` el cual mostrará la información del automóvil deportivo en base a los valores de los **Atributos**, es un **Método** nuevo.
* Un **Método** llamado `.relPesoPot()` el cual calcula la relación entre el peso y la potencia del automóvil deportivo, es un **Método** nuevo.

Cuando se hace la operación de **Herencia**, realmente se intenta hacer una versión de la **Clase** ***padre*** más específica y, por tanto, se pretende aprovechar al máximo el trabajo realizado y definido en la misma. Con el fin de poder reutilizar el código de la **Clase** ***padre***, **Python** define un mecanismo para llamar a las funciones de esa **Clase** y obtener los resultados desde la instancia hijo, esto se hace mediante la función `super()` como se muestra a continuación.

Con ello, se definirá un **Objeto** del tipo `automovilDeportivo` correspondiente a un ***McLaren 750S*** con valores específicos para sus **Atributos** (incluyendo los correspondientes a la **Clase** ***padre***). De esa manera se podrán aplicar los dos **Métodos** nuevos que se agregaron, pero también se podrá hacer uso de los **Métodos** de la **Clase** ***padre***.

```python
#Creación de la Clase
class automovilDeportivo(Automovil):
    def __init__(self, nombre, marca, modelo, motor, llanta, odometro, vel_maxima, potencia, peso):
        super().__init__(nombre, marca, modelo, motor, llanta, odometro)
        self.vel_maxima = vel_maxima
        self.potencia   = potencia
        self.peso       = peso
        
    def mostrarInfoDep(self):
        print("El automóvil %s de la marca %s tiene un motor de %.1f litros y alcanza los %d km/hr con %d hp." 
              %(self.nombre, self.marca, self.motor, self.vel_maxima, self.potencia))
    
    def relPesoPot(self):
        pesopot = self.peso / self.potencia
        print("La relación Peso a Potencia del %s es de %.2f kilogramos por caballo de potencia." 
              %(self.nombre, pesopot))

#Creación del Objeto con sus respectivos Atributos
mclaren = automovilDeportivo('750S', 'McLaren', 2020, 4.0, '305/30R20', 50000, 332, 750, 1277)

#Aplicación del Método para el Objeto
mclaren.mostrarInfoDep()
print("\n")

#Aplicación del Método para el Objeto
mclaren.relPesoPot()
print("\n")

#Aplicación del Método de la Clase padre para el Objeto
mclaren.datosLlanta()
```
