# ALMACENAMIENTO DE DATOS DE SQL A ARCHIVOS CSV

```python
import sqlite3 as sql
connection = sql.connect('ds_salaries.sqlite')

# 2. Creaci√≥n de un objeto cursor
cursor = connection.cursor()

# 3. Se crea una consulta de prueba
q = '''
SELECT *
FROM ds_salaries
'''

# 4. Se ejecuta el query
cursor.execute(q)

# 5. Se obtienen los resultados de la consulta
results = cursor.fetchall()

for r in results:
    print(r)
```

```python
# Crea la consulta
q = '''
SELECT CAST(c2 AS INT) as yir, c9 as ctry, c5 as job_title, CAST(c8 AS FLOAT) as salary
FROM ds_salaries;
'''
# Ejecuta la consulta
cursor.execute(q)
# Obtiene resultados
results = cursor.fetchall()
# Imprime
for r in results:
    print(r)
```

```python
import pandas as pd
df = pd.DataFrame(results)
```

```python
# Agregar los nombres de columnas
col_titles = ['year','country','job_title','salary']
df.columns = col_titles
```
